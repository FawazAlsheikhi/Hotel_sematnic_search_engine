```python
# Returns the names of hotels with missing descriptions
missing_description_hotels = df.loc[df['hotel_description'].isna(), 'hotel_name'].unique()
```

```python
# find the discription of these hotels using OPEN AI CHATGPT API
#!pip3 install openai==0.28
from openai import AsyncOpenAI


OPENAI_API_KEY = "your_api_key"
openai.api_key = OPENAI_API_KEY
client = AsyncOpenAI(api_key = OPENAI_API_KEY)

async def get_gpt_response(message: str) -> str:
    response = await client.chat.completions.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": message}
        ],
        max_tokens=500
    )
    return response.choices[0].message.content

hotel_discription_dic = {hotel:"" for hotel in missing_description_hotels}

for hotel in missing_description_hotels:

    qurey = f"Find me the discription of this hotel :{hotel} if you couldn't find information return only number 0 without anything else"
    retreived_hotel_discription = await get_gpt_response(qurey)
    hotel_discription_dic[hotel] = retreived_hotel_discription
    print(hotel_discription_dic[hotel])
    






```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
def filter_and_rank_by_similarity(query, df, model,cities):
    city = extract_city(query, cities)
    cities = df['locality'].unique()

    if city in cities:
        df = df.loc[df['locality'] == city]
        print(f"City: {city}")
    else:
        print("No city found.")


    query_embedding = model.encode(query)
    embeddings_matrix = np.stack(df['embedding'].values)
    similarities = cosine_similarity([query_embedding], embeddings_matrix).flatten()
    df['similarity'] = similarities
    df = df.loc[df.groupby('hotel_name')['similarity'].idxmax()]
    top_5_similar = df.nlargest(5, 'similarity')

    return top_5_similar
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-classification", model="yiyanghkust/finbert-tone")
```

```python
df['classifcation'] = df.review_text.apply(lambda x: pipe(x))
```

```python
df['classifcation'] = df.classifcation.apply(lambda x: x[0]['label'])
```

```python
classifcation_scores = {"Negative": -1 , "Neutral" : 0, "Positive":1}

df['classifcation_numeric'] = df['classifcation'].map(classifcation_scores)
```

```python
import pandas as pd


# Group by 'hotel_name' and calculate the sum and count
grouped = df.groupby('hotel_name').agg(
    sum_classification=('classifcation_numeric', 'sum'),
    count_rows=('classifcation_numeric', 'size')
)

# Calculate the desired value by dividing sum by (count * 2)
grouped['result'] = grouped['sum_classification'] / (grouped['count_rows'])

# Reset index to make 'hotel_name' a column again
result = grouped.reset_index()
```

```python
def filter_and_rank_by_similarity_sentiment_ranking(query, df, model,cities, k):
    city = extract_city(query, cities)
    cities = df['locality'].unique()
    if city in cities:
        df = df.loc[df['locality'] == city]
        print(f"City: {city}")
    else:
        print("No city found.")


    query_embedding = model.encode(query)
    embeddings_matrix = np.stack(df['embedding'].values)
    similarities = cosine_similarity([query_embedding], embeddings_matrix).flatten()
    df['similarity'] = similarities
    df = df.loc[df.groupby('hotel_name')['similarity'].idxmax()]

    top_5_similar = df.nlargest(k, 'similarity').sort_values(by=['result'], ascending=False)
    return top_5_similar
```

```python
#find the most similar based on cosine smiliarty,
#then sort them based on the score of the sentiment analysis of reviews

query = 'close to the Effiel Tower and very high in Paris'
query_output = filter_and_rank_by_similarity_sentiment_ranking(query,df,model, cities, 20)
query_output[['hotel_name',"similarity","rate",'locality',"result"]].head(20)
```

```python
import gradio as gr

def gr_request(query):
    result = filter_and_rank_by_similarity_sentiment_ranking(query ,df, model, cities, 3)
    #print(result.head(5))
    result = result[['hotel_name','locality','similarity','result','rate','review_text',"hotel_image"]].values
    result_images = [res[6] for res in result]
    result_texts = [f"Hotel name: {hotel[0]} \nCity: {hotel[1]}\nSimilarity score: {round(hotel[2],2)} \nSentiment score: {round(hotel[3],2)} \nRate: {hotel[4]}/5.0 \n\n\nBest Match Review:\n{hotel[5]}" for hotel in result]

    output_list = list()
    for i in range(len(result_images)):
        output_list.append(result_images[i])
        output_list.append(result_texts[i])
    return output_list


demo = gr.Interface(fn=gr_request,inputs = ["text"], outputs=['image',"text",'image', "text",'image', "text"])
if __name__ == "__main__":
    demo.launch(share=True)
```
